```python
import mysql.connector as conn
import pandas as pd
import csv

try:
    db = conn.connect(host = 'localhost', database = "Rajshekhar1998", user = 'root', passwd = 'root')
    cur = db.cursor()
    query = "CREATE TABLE IF NOT EXISTS IndustryTypes (Subpart_letter VARCHAR(20), Name_Of_Industry VARCHAR(50), Facility_Type VARCHAR(30))"
    cur.execute(query)
    with open("Industry Types.csv", 'r') as data:
        next(data)
        data_csv = csv.reader(data, delimiter = ",")
        for line in data_csv:
            print(line)
            #for list in line[1]: 
            print(tuple(line))
            cur.execute('insert into IndustryTypes values {data}'.format(data = tuple(line)))
        print("all data inserted")
    db.commit()
except Exception as e:
    print(str(e))
```
